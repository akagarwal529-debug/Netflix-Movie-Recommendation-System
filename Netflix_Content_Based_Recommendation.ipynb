```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv('netflix_titles.csv')
```

```python
features = ['title', 'director', 'cast', 'listed_in', 'description']
df = df[features].copy()
```

```python
df.fillna('', inplace=True)
```

```python
df['combined_features'] = (
    df['listed_in'] + ' ' +
    df['cast'] + ' ' +
    df['director'] + ' ' +
    df['description']
)
```

```python
df[['title', 'combined_features']].head()
```

```python
df['combined_features'] = df['combined_features'].str.lower()
```

```python
tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(df['combined_features'])
```

```python
tfidf_matrix.shape
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim.shape
```

```python
df = df.reset_index(drop=True)
```

```python
def recommend_movies(movie_title, num_recommendations=5):
    # Get index of the movie
    idx = df[df['title'] == movie_title].index[0]
    
    # Get similarity scores
    sim_scores = list(enumerate(cosine_sim[idx]))
    
    # Sort movies based on similarity score
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Get top N similar movies (excluding itself)
    sim_scores = sim_scores[1:num_recommendations+1]
    
    # Get movie titles
    movie_indices = [i[0] for i in sim_scores]
    
    return df['title'].iloc[movie_indices]
```

```python
recommend_movies("Inception")
```

```python
recommend_movies("Breaking Bad")
```

```python
import pickle
```

```python
model_data = {
    "tfidf_vectorizer": tfidf,
    "tfidf_matrix": tfidf_matrix,
    "cosine_similarity": cosine_sim,
    "dataframe": df
}
```

```python
with open("netflix_recommendation_model.pkl", "wb") as file:
    pickle.dump(model_data, file)

print("Model saved successfully as netflix_recommendation_model.pkl")
```
